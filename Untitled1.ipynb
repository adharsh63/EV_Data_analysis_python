```python
import pandas as pd

# This loads your file into a variable called 'df'
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')

# This will display the first 5 rows to confirm it worked
df.head()
```

```python
print("Missing values per column:")
print(df.isnull().sum())
df = df.drop_duplicates()
df['Legislative District'] = df['Legislative District'].fillna(0)
print("\nCleaning complete! New shape of data:", df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Electric Range'])
plt.title('Checking for Outliers in Electric Range')
plt.show()
```

```python
top_10_makes = df['Make'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_makes.index, y=top_10_makes.values, palette='magma')
plt.xticks(rotation=45)
plt.xlabel('Car Brand (Make)')
plt.ylabel('Number of Vehicles')
plt.title('Top 10 Electric Vehicle Brands in the Dataset')
plt.show()
```
